```python
def is_inf(point):
    #Check if the point is the point of infinity
    #input: - point from E_{A,B}
    #output: true if the point is the point of inf and false otherwise
    if(len(point)==3 and point[2]==0):
        return true
    return false
```

```python
def reverse_point(point):
    #Return the inverse of a point
    #input: a point (x,y,1) from E_{A,B}
    #output: the reverse point (x,-y,1)
    x = point[0]
    y = point[1]
    return (x,neg_fp2(y,p),1)
```

```python
def is_inverse(point1, point2):
    # Check if a point 1 is an inverse to point 2
    #input: -point1, point2; two points from E
    #output: True if point1 is an inverse to point2 and false otherwise
    x1 = point1[0]
    y1 = point1[1]
    x2 = point2[0]
    y2 = point2[1]
    if (point1 == reverse_point(point2)):
        return true
    return false
```

##  coordinates doubling

```python
#### This function is updated
def xDBL(point, a, b):
    #point doubling function
    #Input: - a point from E_{a,b}
    # a,b: coefficients to define E_{a,b}
    #output: (x_[2p], y_[2p]) doubling point 
    if(is_inf(point)): return point
    xp = point[0] # in FP^2
    yp = point[1] # in FP^2
    t0 = square_fp2(xp,p) #xp^2
    t1 = add_fp2(t0,t0,p)#t0+t0 #2x^2
    t0 = add_fp2(t0,t1,p) #t0+t1 #3x^2
    t0 = add_fp2(t0,a,p)#t0+a # 3x^2+a
    t1 = add_fp2(yp,yp,p) #yp+yp # 2.yp
    t1 = invert_fp2(t1,p)#t1^-1 # 
    t1 = multiply_fp2(t0,t1,p)#t0*t1 #** lambda
    #---------------------
    x2p = square_fp2(t1,p) #t1^2
    t0 = add_fp2(xp,xp,p)#xp+xp
    x2p =sub_fp2(x2p,t0,p) #x2p-t0
    #----------------------
    y2p = sub_fp2(xp,x2p,p) #xp-x2p
    y2p =multiply_fp2(t1,y2p,p) #t1*y2p
    y2p = sub_fp2(y2p,yp,p) #y2p-yp
    return (x2p,y2p,1)
```

# Repeated  coordinate doubling

```python
def xDBLe(point, a,b,e):
    # This function calculate [2^e].P by doubling the point e times
    #input:-a point from E
    #      -a,b; the needed coefficients to define E_{a,b} 
    #output: [2^e]P
    point_bar = point
    for iterator in range(1,e+1,1):
        point_bar = xDBL(point_bar,a,b)    
    return point_bar
    
```

# Coordinate addition


```python
def xADD(P, Q,a,b):
    #works successfully
    # This functions add two points from E_{a,b}
    # input: -P,Q from E_{a,b}
    #        -a,b the needed coefficients to define E_{a,b}
    # output: P+Q
    xP = P[0] # in FP^2
    yP = P[1] # in FP^2
    xQ = Q[0] #in FP^2
    yQ = Q[1] #in FP^2
    if(is_inverse(P,Q)):
        return inf_point  ## we return a point not x,y because of this thr point of inf (I put this first in the case of calling trippling I was getting error)
    if (is_inf(P)):
        return Q
    if (is_inf(Q)):
        return P
    if(P==Q):
        return xDBL(P,a,b)
   
    t0 = sub_fp2(yQ,yP,p) #yQ-yP
    t1 = sub_fp2(xQ,xP,p)# xQ-xP
    t1 = invert_fp2(t1,p)#t1^-1
    t0 = multiply_fp2(t0,t1,p) #t0*t1 # lambda
    t1 = square_fp2(t0,p) #t0^2 # lambda^2
    t2 = add_fp2(xP,xP,p) #xP+xP # 2xP
    t2 =add_fp2(t2,xQ,p) #t2+xQ #2xP+xQ
    #t2 = t2+a # 2xP+xQ+A
    t2 = multiply_fp2(t2,t0,p) #t2*t0#(2xP+xQ)*lambda
    t0 = multiply_fp2(t0,t1,p) #t0*t1 # lambda^3
    #t0 = b*t0 # b*lambda^3
    t0 = add_fp2(t0,yP,p) #t0+yP # lambda^3 + yp
    t0 = sub_fp2(t2,t0,p) #t2-t0 # (2xP+xQ)*lambda - lambda^3-yp
    #t1 = b*t1 # b*lambda^2
    #t1 = t1-a # b*lambda^2-a
    t1 = sub_fp2(t1,xP,p) #t1-xP #lambda^2 -xP
    xpq = sub_fp2(t1,xQ,p) #t1-xQ # lambda^2 -xP -xq
    ypq = t0 #(2xP+xQ)*lambda - lambda^3-yp
    return (xpq,ypq,1)    
      
```

#  coordinates tripling

```python
def xTPL(p,a,b):
    # This function tripple a point P from E by first doubling the point then add the result to P
    #input: -P a point from E_{a,b}
    #       -a,b: the needed cofficients to define E_{a,b}
    double_point = xDBL(p,a,b) 
    Point = xADD(p,double_point,a,b)
    return Point
```

# Repeated affine coordinate tripling


```python
def xTPLe(point, a,b,e):
    # This function calculate [3^e].P by trippling the point e times
    #input:-a point from E
    #      -a,b; the needed coefficients to define E_{a,b} 
    #output: [3^e]P
    point_bar = point
    for iterator in range(1,e+1,1):
        point_bar = xTPL(point_bar,a,b)  
    return point_bar
```

# Double and add scalar multiplication


```python
def double_and_add(m, point,a,b):
    # This function calculates the scalar multiplication of a point (i.e. [m].P)
    # input: - m a scalar value
    #        - point: a point from E_{a,b}
    #        - a,b; the needed coefficients to define E_{a,b}
    # output: [m].P(i.e. the scalar multiplication of P by m)
    point0 = inf_point
    binrep = mpz(m).digits(2) # binary representation of m
    length = len(binrep)
    for iterator in range(0,length,1):
        point0 = xDBL(point0,a,b)
        if(binrep[iterator]=='1'):
            point0 = xADD(point0, point, a,b)
    return point0
```

```python
def j_inv(a,b):
    #This function calculates the J-invaraint of an elliptic curve 
    #input: -a the coefficient that define the elliptic curve E_{a,b}
    #output: j(E_{a,b}) = (1728*(4.a^3)/(4.a^3+27b^2))
    #a= a + 0*i
    #b= b+0*i
    
    t0 = square_fp2(a,p)  #a^2
    t0 = multiply_fp2(t0,a,p)#t0*a #a^3
    t0 = add_fp2(t0,t0,p) #t0+t0 # 2a^3
    t0 = add_fp2(t0,t0,p) #t0+t0 # 4a^3 ***
    t1 = square_fp2(b,p)# b^2 #****
    t2 = add_fp2(t1,t1,p) #t1+t1 # 2b^2 ***
    t3 = add_fp2(t2,t2,p) #t2+t2 #4
    t3 = add_fp2(t3,t3,p) #t3+t3 #8 *****
    t4 = add_fp2(t3,t3,p) #t3+t3 #16 ****
    t4 = add_fp2(t4,t3,p) #t4+t3 #24
    t4 = add_fp2(t4,t2,p) #t4+t2 #26
    t4 = add_fp2(t4,t1,p) #t4+t1 #27*b^2

    t4 = add_fp2(t4,t0,p) #t4 + t0 # 4.a^3+27.b^2
    t4 = invert_fp2(t4,p) #t4^-1
    t4 = multiply_fp2(t4,t0,p) #t4*t0
    t4 = add_fp2(t4,t4,p) #t4+t4 # 2
    t4 = add_fp2(t4,t4,p)#t4+t4 # 4
    t4 = add_fp2(t4,t4,p) #t4+t4 #8
    t4 = add_fp2(t4,t4,p) #t4+t4 #16
    t4 = add_fp2(t4,t4,p) #t4+t4 #32
    t4 = add_fp2(t4,t4,p) #t4+t4 #64 **
    t0 = add_fp2(t4,t4,p) #t4+t4 #128 *
    t1 = add_fp2(t0,t0,p) #t0+t0 #256
    t1 = add_fp2(t1,t1,p) #t1+t1 #512*
    j = add_fp2(t1,t1,p) #t1+t1 #1024*
    j = add_fp2(j,t1,p)#j+t1 # 1536
    j = add_fp2(j,t0,p)#j+t0 # 1664
    j = add_fp2(j,t4,p) #j+t4 #1728
    return j
```

# Computing the 2_isogenous curve


```python
def curve_2_iso(x,a,b):
    #This function computes a 2-isogeny map (phi_2:(E_{a,b}) --->(E'_{a',b'}))based on a cyclic-  
    ##subgroup generated by point (x,y), the order of the point (x,y) is exactly 2
    # input: -x, the x-coordinate of a point P(x,y) with order =2
    #        -b the coefficient that define the elliptic curve E_{a,b}
    # output: -a_bar, b_bar that defines E_{a_bar, b_bar} (i.e. the co-domain elliptic curve)
    #a = a+ 0*i
    #b = b+ 0*i
    t0 = square_fp2(x,p) #x^2
    t1 = add_fp2(t0,t0,p) #t0+t0
    t1 = add_fp2(t1,t0,p) #t1+t0 #3x^2
    t1 = add_fp2(t1,a,p) #t1+a #3.x^2+a
    t2 = add_fp2(t1,t1,p) #t1+t1 #2 (3x^2+a)
    t2 = add_fp2(t2,t2,p) #t2+t2 # 4(3x^2+a)
    t2 = add_fp2(t2,t1,p) #t2+t1 # 5(3x^2+a)
    w = multiply_fp2(x,t1,p) #x*t1 #x (3x^2+a)
    w1 = add_fp2(w,w,p)# w+w #2w
    w2 = add_fp2(w1,w1,p) #w1+w1 #4w
    w2 = add_fp2(w2,w1,p) #w2+w1 #6w1
    w2 = add_fp2(w2,w,p) #w2+w #7w1
    a_bar = sub_fp2(a,t2,p) #a -t2
    b_bar = sub_fp2(b,w2,p) #b-w2
    return a_bar, b_bar, t1 #t1 = 3.xp^2+a
    
```

# Evaluating a 2-Isogeny at a point


```python
def eval_2_iso(point, xp2,a,K):
    #This function evaluates a point from E according to the isogeny map of degree-2(computed based on the kernel <(xp2, yp2)>) 
    # input: - a point from E_{a,b}
    #        - xp2; the x-coordinte of the point (xp_2, yp_2) of degree 2
    #        -a: the cofficient of the elliptic curve E_{a,b}
    #        -K : a pre calculated value during curve-2-iso = 3xp^2+a to save calculations
    # output: (xpoint_bar, ypoint_bar) = phi_2(point) = phi_2(x,y)
    # xpoint_bar = x + (3.{x_{P2}}^2+a) / (x-x_{P2})
    # ypoint_bar = y - (y(3.{x_{P2}}^2+a)/(x-x_{P2})^2)
    x = point[0]
    y = point[1]
    #------------------------
    #t0 = square_fp2(xp2,p) #xp2^2
    #t1 = add_fp2(t0,t0,p) #t0+t0 #2xp2^2
    #t1 = add_fp2(t1,t0,p) #t1+t0 # 3xp2^2
    #t1 = add_fp2(t1,a,p) #t1+a # 3xp2^2+a **save 
    
    t2 = sub_fp2(x,xp2,p) #x-xp2
    xpoint_bar = invert_fp2(t2,p) #t2^-1 #(x-xp2)^-1
    t1 = square_fp2(xpoint_bar,p) #t2^{-2}
    xpoint_bar = multiply_fp2(K,xpoint_bar,p) #t1* xpoint_bar # 
    xpoint_bar = add_fp2(x,xpoint_bar,p) #x+xpoint_bar
    #------------------------
    ypoint_bar = multiply_fp2(y,K,p) #y*t1
    t1 = multiply_fp2(ypoint_bar,t1,p) #ypoint_bar*t1
    ypoint_bar = sub_fp2(y,t1,p) #y-t1
    #3A+3M+1S+1I
    return (xpoint_bar, ypoint_bar,1)
```

# Computing the 4-isogenous curve

```python
def curve_4_iso(point,a,b):
    #This function computes a 4-isogeny map to accelerate execution (phi_4:(E_{a,b}) --->(E'_{a',b'}))based on a cyclic-  
    ##subgroup generated by point (xp,yp), the order of the point (xp,yp) is exactly 4
    # input: - a point P(x,y) with order =4
    #        -a,b the coefficient that define the elliptic curve E_{a,b}
    # output: -a_bar, b_bar that defines E_{a_bar, b_bar} (i.e. the co-domain elliptic curve)
    xp = point[0]
    yp = point[1]
    
    g0x = square_fp2(xp,p) #xp^2
    g1x = add_fp2(g0x,g0x,p) #g0x+g0x # 2xp^2
    g1x = add_fp2(g1x,g0x,p) #g1x+g0x # 3xp^2
    g1x = add_fp2(g1x,a,p) #g1x+a # 3xp^2+a **
    v1x = add_fp2(g1x,g1x,p) #g1x+g1x #2g1x ***
    
    
    t0 = yp
    t0 = add_fp2(t0,t0,p) #t0+t0 #2yp
    t0 = invert_fp2(t0,p) #t0^-1 #(2yp)^-1
    lamda =multiply_fp2(g1x,t0,p) #g1x*t0 #(2yp)^-1*(3xp^2+a) *** lambda
    x2 = square_fp2(lamda,p) #lamda^2
    t0 = xp
    t0 = add_fp2(t0,t0,p) #t0+t0 #2xp
    x2 = sub_fp2(x2,t0,p) #x2-t0 #(4yp^2)^-1*(3xp^2+a)^2 - 2xp
    g3x = square_fp2(x2,p)#x2^2
    g2x = add_fp2(g3x,g3x,p)#g3x+g3x 
    g2x = add_fp2(g2x,g3x,p) #g2x+g3x
    g2x = add_fp2(g2x,a,p) #g2x+a # *** save
    
    v2x = g2x #***save importnat vp2 = g2x because this point with order =2
    v0 = add_fp2(v1x,v2x,p) #v1x+v2x
    v = add_fp2(v0,v0,p)#v0+v0 #2v0
    v = add_fp2(v,v,p)#v+v #4v0
    v = add_fp2(v,v0,p) #v+v0 # 5v0
    a_bar = sub_fp2(a,v,p) #a-v
    #-----------------------------------------------------------
    #y2 = g2y = y2y =0 (always =0 since the point 0f y2 has order =2)
    ##### g1y = initit_fp2mpz(-1)
    g1y = neg_fp2(yp,p)
    ### g1y = multiply_fp2(g1y,yp,p) #g1y*yp
    g1y = add_fp2(g1y,g1y,p) #g1y+g1y # -2yp
    u1y = square_fp2(g1y,p) #g1y^2
    
    w0 =multiply_fp2(xp,v1x,p) #xp*v1x
    w0 = add_fp2(u1y,w0,p)#u1y+w0
    t0 = multiply_fp2(x2,v2x,p)#x2*v2x
    w0 = add_fp2(w0,t0,p)#w0+t0
    w1 = add_fp2(w0,w0,p) #w0+w0 # 2w0
    w = add_fp2(w1,w1,p) #w1+w1 #4w0
    w = add_fp2(w,w1,p) #w+w1 #6w0
    w = add_fp2(w,w0,p) #w+w0 #7w0
    
    b_bar = sub_fp2(b,w,p) #b-w
    
    return a_bar, b_bar,x2, g1x, g2x, g1y, v1x, v2x, u1y # extra parameteres are used in the eval function to avoid doing the calculation again
    
    
```

# Evaluating a 4-isogeny at a point

```python
def eval_4_iso(point,point4,x2, g1x, g2x, g1y,v1x,v2x,u1y):
    #This function evaluates a point from E according to the isogeny map of degree-4(computed based on the kernel <(xp4, yp4)>) 
    # input: - a point from E_{a,b}
    #        - point4 : a point (xp4, yp4) of degree 4
    # output: (xQ_bar, yQ_bar) = phi_4(point) = phi_4(xQ,yQ)
    x = point[0]
    y = point[1]
    
    xp = point4[0]
    yp = point4[1]
    
    t0 = sub_fp2(x,xp,p) # x-x4 ** save for y
    t1 = multiply_fp2(t0,v1x,p) # v1x(x-x4)
    t1 = add_fp2(t1,u1y,p) # v1x(x-x4)+u1y 
    t2 = sub_fp2(x,x2,p) # x-x2
    t3 = square_fp2(t0,p) #(x-x4)^2
    
    t4 = multiply_fp2(t0,t3,p) # (x-x4)^3 save for y **
    t5 = square_fp2(t2,p) # (x-x2)^2 save for y **
    
    t6 = multiply_fp2(v2x,t3,p) #v2x *(x-x4)^2 
    t3 = multiply_fp2(t3,t2,p) #(x-x4)^2 (x-x2) **
    t1 = multiply_fp2(t1,t2,p) #[v1x(x-x4)+u1y](x-x2) 
    t1 = add_fp2(t1,t6,p) # [v1x(x-x4)+u1y](x-x2) + v2x*(x-x4)^2  **
    t3 = invert_fp2(t3,p) # [(x-x4)^2 (x-x2)]^-1
    t3 = multiply_fp2(t1,t3,p) # [v1x(x-x4)+u1y](x-x2) + v2x*(x-x4)^2 *  [(x-x4)^2 (x-x2)]^-1
    x_bar = add_fp2(x,t3,p) # 5A+6M+2S+1I
    
    t1 = add_fp2(u1y,u1y,p) # 2u1y
    t1 = multiply_fp2(t1,y,p) # 2u1y*y
    t2 = sub_fp2(y,yp,p) # y-y4
    t2 = multiply_fp2(v1x,t2,p) # v1x(y-y4)
    t3 = multiply_fp2(g1x,g1y,p) # g1x*g1y
    t2 = sub_fp2(t2,t3,p) # v1x(y-y4) - g1x*g1y
    t2 = multiply_fp2(t2,t0,p) # [v1x(y-y4) - g1x*g1y]*(x-x4)
    t2 = add_fp2(t1,t2,p) # 2u1y*y+  [v1x(y-y4) - g1x*g1y]*(x-x4)
    t2 = multiply_fp2(t5,t2,p) #  (x-x2)^2 *[2u1y*y+  [v1x(y-y4) - g1x*g1y]*(x-x4)]
    t1 = multiply_fp2(v2x,y,p) # v2x*y
    t1 = multiply_fp2(t1,t4,p) # v2x*y * (x-x4)^3
    t2 = add_fp2(t2,t1,p) # (x-x2)^2 *[2u1y*y+  [v1x(y-y4) - g1x*g1y]*(x-x4)]+v2x*y * (x-x4)^3
    t1 = multiply_fp2(t4,t5,p) # (x-x4)^3 * (x-x2)^2
    t1 = invert_fp2(t1,p) # [ (x-x4)^3 * (x-x2)^2]^-1
    t2 = multiply_fp2(t2,t1,p) #{(x-x2)^2 *[2u1y*y+  [v1x(y-y4) - g1x*g1y]*(x-x4)]+v2x*y * (x-x4)^3 }*[ (x-x4)^3 * (x-x2)^2]^-1
    y_bar = sub_fp2(y,t2,p) # y-t2
    # 6A+9M+1I
    ##########################################
    
    return (x_bar, y_bar,1)
```

# Computing the 3-isognenous curve

```python
def curve_3_iso(point,a,b):
    #This function computes a 3-isogeny map (phi_3:(E_{a,b}) --->(E'_{a',b'}))based on a cyclic-  
    ##subgroup generated by point (x,y), the order of the point (x,y) is exactly 3
    # input: -point,  a point P(x,y) with order =3
    #        -a,b the coefficient that define the elliptic curve E_{a,b}
    # output: -a_bar, b_bar that defines E_{a_bar, b_bar} (i.e. the co-domain elliptic curve)
    # a_bar = a - 10 (3.x^2+a)
    # b_bar = b - 28y^2-14x(3x^2+a)
    x = point[0]
    y = point[1]
    #### a = a+ 0*i
    ### b = b+ 0*i
    #-------------------------
    t0 = square_fp2(x,p) #x^2
    t1 = add_fp2(t0,t0,p) #t0+t0 #2x^2
    t1 = add_fp2(t1,t0,p) #t1+t0 #3x^2
    t1 = add_fp2(t1,a,p) #t1+a #3.x^2+a ***
    K = t1
    t2 = add_fp2(t1,t1,p) #t1+t1 #2 (3x^2+a) ***
    t3 = add_fp2(t2,t2,p) #t2+t2 # 4(2x^2+A)
    t0 = add_fp2(t3,t3,p) #t3+t3 # 8(3x^2+a)
    t0 = add_fp2(t2,t0,p)#t2+t0 # 10(3x^2+a)
    a_bar = sub_fp2(a,t0,p) #a-t0 # a-10(3x^2+a)
    
    t3 = add_fp2(t3,t0,p) #t3+t0 # 14(3x^2+a)
    t3 = multiply_fp2(t3,x,p) #t3*x # 14*x*(3x^2+a) **
    t0 = square_fp2(y,p) #y^2
    t0 = add_fp2(t0,t0,p) #t0+t0 # 2*y^2
    t0 = add_fp2(t0,t0,p) #t0+t0 # 4*y^2
    t1 = add_fp2(t0,t0,p)# t0+t0 # 8*y^2
    t2 = add_fp2(t1,t1,p) #t1+t1 # 16*y^2
    t2 = add_fp2(t2,t1,p) #t2+t1 #24*y^2
    t2 = add_fp2(t2,t0,p) #t2+t0 #28*y^2
    b_bar = sub_fp2(b,t2,p) #b-t2 # b-28*y^2
    b_bar = sub_fp2(b_bar,t3,p)# b_bar - t3 # b-28*y^2 - 14*x*(3x^2+a)
    #17A+1M+2S
    return a_bar, b_bar, K 
```

# Evaluating a 3-isogeny at a point

```python
def eval_3_iso(point, point3,a, K):
    #This function evaluates a point from E according to the isogeny map of degree-3(computed based on the kernel <(xp3, yp3)>) 
    # input: - a point from E_{a,b}
    #        - point3; a point (xp3, yp3) from E_{a,b} of degree 3
    #        -a: the cofficient of the elliptic curve E_{a,b}
    #        -K : a constant that makes calculation faster
    # output: (xpoint_bar, ypoint_bar) = phi_2(point) = phi_2(x,y)
    # xpoint_bar =  x + (2(3(xp3)^2+a)/(x-xp3) + (4*(yp3)^2)/(x-xp3)^2
    # ypoint_bar = y - (8*y*(yp3)^2)/(x-(xp3)^3) -(2y(3xp3^2+a))/(x-xp3)^2
    x = point[0]
    y = point[1]
    xp3 = point3[0]
    yp3 = point3[1]
    #------------------------

    t1 =K
    t1 = add_fp2(t1,t1,p) #t1+t1 # 2(3xp^2+a)
    t4 = multiply_fp2(t1,y,p) #t1*y #2*y*(3xp^2+a)
    t0 = sub_fp2(x,xp3,p) #x-xp3 #*** x-xp3
    t2 = invert_fp2(t0,p) #t0^-1 # (x-xp3)^-1
    t3 = square_fp2(t2,p)#((x-xp3)^2)^-1
    t0 = multiply_fp2(t2,t3,p)#((x-xp3)^2)^-3
    t2 = multiply_fp2(t1,t2,p) #t1*t2 # 2(3xp3^2+a)*(x-xp3)^-1 ***
    t1 = square_fp2(yp3,p) #yp3^2
    t1 = add_fp2(t1,t1,p) #t1+t1 # 2(yp3)^2
    t1 = add_fp2(t1,t1,p) #t1+t1 # 4(yp3)^2 ***
    t2 = add_fp2(x,t2,p) #x+t2 # x+2(3xp^3+a)*(x-xp3)^-1
    xpoint_bar =multiply_fp2(t1,t3,p) #t1*t3 # 4(yp3)^2*((x-xp3)^2)^-1
    xpoint_bar = add_fp2(t2,xpoint_bar,p) #t2+xpoint_bar #x+2(3xp3^2+a)*(x-xp3)^-1 +4(yp3)^2*((x-xp3)^2)^-1
    #---------------------------------------------------------------------------------------
    t2 = add_fp2(t1,t1,p) #t1+t1 # 8(yp3)^2
    t2 =multiply_fp2(y,t2,p) #y*t2 # 8*y*(yp3)^2
    t1 = multiply_fp2(t4,t3,p) #t4*t3 # 2y(2xp3+a)*((x-xp3)^2)^-1
    t0 = multiply_fp2(t2,t0,p) #t2*t0 # 8*y*(yp3)^2 * ((x-xp3)^3)^-1
    ypoint_bar = sub_fp2(y,t0,p) #y-t0 # y-8*y*(yp3)^2 * ((x-xp3)^3)^-1
    ypoint_bar = sub_fp2(ypoint_bar,t1,p) #ypoint_bar-t1 ## y-8*y*(yp3)^2 * ((x-xp3)^3)^-1 - 2y(2xp3+a)*((x-xp3)^2)^-1
    #9A+7M+2S+1I
    return (xpoint_bar, ypoint_bar,1)
```

# Computing and evaluating a 2^e isogeny, simple version

```python
def iso_2_e(a,b, SA, BobPoints=[]):
    # This function computes the isogeny map of Alice of degree 2^eA by decompsing this map into e isogeny maps of degree-2
    #input: - -a,b the coefficient that define the elliptic curve E_{a,b}
    #       - SA is the secret point of Alice with orderof 2^eA
    #       -BobPoints: Bob's points to be evaluated according to Alice's map
    #        we have two casess here:
    #              (a)When Alice is caluclating her public key she needs to evaluate Bob's points according to her map.
    #              (b) When Alice is calculating her shared key, no need to evaluate the points (the list is empty)
    #output:
    #        - a_bar, b_bar that defines E_{a_bar, b_bar} (i.e. the co-domain elliptic curve)
    #        - BobPoints after evaluation according to the isogeny map of degree 2^eA
    a_bar = a
    b_bar = b 
    e2_bar = eA
    
    if(e2_bar%2!=0):
        point_T = xDBLe(SA, a,b,e2_bar-1) ## a point with order 2
        a_bar, b_bar,K = curve_2_iso(point_T[0],a,b) ## x of point_T with order 2
        SA = eval_2_iso(SA,point_T[0],a,K)
        for iterator in range(len(BobPoints)):
            BobPoints[iterator] = eval_2_iso(BobPoints[iterator], point_T[0],a,K)
        e2_bar = e2_bar-1
      
    for e in range(e2_bar-2,-1,-2):
        point_T = xDBLe(SA, a_bar, b_bar, e) ## the point with order =4
        a_bar, b_bar,x2, g1x, g2x, g1y,v1x,v2x,u1y = curve_4_iso(point_T,a_bar,b_bar)
        if(e!=0):
            SA = eval_4_iso(SA, point_T,x2, g1x, g2x, g1y,v1x,v2x,u1y)
        
        for iterator in range(len(BobPoints)):
            BobPoints[iterator] = eval_4_iso(BobPoints[iterator], point_T,x2, g1x, g2x, g1y,v1x,v2x,u1y)
    return a_bar, b_bar, BobPoints
        
```

# Computing and evaluating a 3^e isogeny , simple version


```python
def iso_3_e(a,b,SB, AlicePoints=[]):
    # This function computes the isogeny map of Bbo of degree 3^eB by decompsing this map into e isogeny maps of degree-3
    #input: - -a,b the coefficient that define the elliptic curve E_{a,b}
    #       - SB is the secret point of Alice with orderof 3^eB
    #       -AlicePoints: Alice's points to be evaluated according to Bob's map
    #        we have two casess here:
    #              (a) When Bbo is caluclating his public key, he needs to evaluate Alice's points according to his map.
    #              (b) When Bob is calculating his shared key, no need to evaluate the points (the list is empty)
    #output:
    #        - a_bar, b_bar that defines E_{a_bar, b_bar} (i.e. the co-domain elliptic curve)
    #        - AlicePoints after evaluation according to the isogeny map of degree 3^eB
    #print("SB before evaluation: ", SB)
    #a_bar = a
    #b_bar = b
    for e in range(eB-1,-1,-1):
        point_T = xTPLe(SB, a, b, e)
        #print("point with order 3: ", point_T)
        a_bar, b_bar,K = curve_3_iso(point_T,a,b)
        #print("here: ", "a_bar", a_bar," ", "b_bar = ", b_bar)
        #print()
        if(e!=0):
            SB = eval_3_iso(SB,point_T,a,K)
            #print("SB evaluation: ", SB)
        for iterator in range(len(AlicePoints)):
            #print("a = ", a)
            #print("point = ", AlicePoints[iterator])
            #print("point 3 =", point_T)
            AlicePoints[iterator] = eval_3_iso(AlicePoints[iterator], point_T,a,K)
            #print("Alice's points evaluation: ", AlicePoints[iterator])
        a = a_bar
        b = b_bar    
    return a_bar, b_bar, AlicePoints
```

#  Computing public key in the 2-torsion

```python
def isogen2(sk2):
    #This function computes the public key of Alice
    #input:  -sk2: a secret scalar value of Alice in the range [0,2^eA)
    #output: -a_bar_Alicepk, b_bar_Alicepk defining the elliptic curve E_{a_bar_Alicepk, b_bar_Alicepk} (i.e, the public elliptic curve of Alice)
    #        - BobPoints: evaluated according to Alice's map
    
    ## a, b have their initial values
    ## PA, QA have initial values
    ## BobPoints [PB,QB] have initial values
    SA = double_and_add(sk2,QA,a,b) ## (scalar multiplication by sk2) the point QA has the order 2^eA
    SA = xADD(PA,SA,a,b) ## The ouput is Alice's secret point: SA = PA + [sk2].QA
    a_bar_Alicepk, b_bar_Alicepk, BobPoints_pk = iso_2_e(a,b,SA,BobPoints)
    return a_bar_Alicepk, b_bar_Alicepk, BobPoints_pk
    
```

# Computing public key in 3-torsion

```python
def isogen3(sk3):
    #This function computes the public key of Bob
    #input:  -sk3: a secret scalar value of Alice in the range [0,3^eB)
    #output: -a_bar_Bobpk, b_bar_Bobpk defining the elliptic curve E_{a_bar_Bobpk, b_bar_Bobpk} (i.e, the public elliptic curve of Bob)
    #        - AlicePoints: evaluated according to Bob's map
    
    ## a,b take their initial values
    ## PB, QB have initial values
    ## Alice Points [PA, QA] have initial values
    SB = double_and_add(sk3,QB,a,b) ## (scalar multiplication by sk3) the point QB has the order 3^eB
    SB = xADD(PB, SB,a,b) ## The ouput is bob's secret point: SB = PB + [sk3].QB
    #print("inside isogen3: ", AlicePoints)
    a_bar_Bobpk, b_bar_Bobpk,AlicePoints_pk = iso_3_e(a,b,SB,AlicePoints)
    
    return a_bar_Bobpk, b_bar_Bobpk, AlicePoints_pk
```

# Establishing shared keys in the 2-torsion


```python
def isox2(sk2, a_Bob_pk , b_Bob_pk, AlicePoitsfromB ):
    # This function computes the shared key of Alice
    #input: -sk2 a secret scalar value of Alice
    #       -a_Bob_pk , b_Bob_pk defining the elliptic curve E_{a_Bob_pk, b_Bob_pk} (i.e, the public elliptic curve of Bob)
    #       -AlicePoitsfromB: Alice's points evaluated according to Bob's map
    #print("inside isox2: ","sk2: ", sk2, "a_Bob_pk: ", a_Bob_pk, "b_Bob_pk: ", b_Bob_pk," ", "Alice point from B [0]: ", AlicePointsfromB[0]," ", "Alice point from B[1]: ", AlicePointsfromB[1] )
    SA_bar = double_and_add(sk2,AlicePoitsfromB[1], a_Bob_pk,b_Bob_pk)
    SA_bar = xADD(AlicePoitsfromB[0], SA_bar, a_Bob_pk,b_Bob_pk)
    #print("The secret point: ", SA_bar)
    a_Bob_pk, b_Bob_pk, _ = iso_2_e(a_Bob_pk,b_Bob_pk,SA_bar)
    Alice_j = j_inv(a_Bob_pk, b_Bob_pk)
    return Alice_j
```

# Establishing shared key in the 3-torsion


## Bob public key generating

```python
def isox3(sk3, a_Alice_pk , b_Alice_pk, BobPointsfromA ):
    # This function computes the shared key of Bob
    #input: -sk3 a secret scalar value of Bob
    #       -a_Alice_pk , b_Alice_pk defining the elliptic curve E_{a_Alice_pk , b_Alice_pk} (i.e, the public elliptic curve of Alice)
    #       -BobPointsfromA: Bob's points evaluated according to Alice's map
    SB_bar = double_and_add(sk3,BobPointsfromA[1], a_Alice_pk,b_Alice_pk)
    SB_bar = xADD(BobPointsfromA[0], SB_bar, a_Alice_pk,b_Alice_pk)
    a_Alice_pk, b_Alice_pk, _ = iso_3_e(a_Alice_pk,b_Alice_pk,SB_bar)
    Bob_j = j_inv(a_Alice_pk, b_Alice_pk)
    return Bob_j
```

## This part measures the times  over all the levels of security

```python
# Define the number of iteration that according to it, we are going to run the measurments
iteration = 5
iteration.save('iteration')
```

```python
## Public Parameters: 
eA_list = [ 216, 250, 305, 372]
eB_list = [137, 159, 192,239]
prime = [434, 503, 610,751]
###----------------------------------------------------------------------------

for index in range(0,iteration,1):
    Alice_time_pk1 = [0,0,0,0]
    Bob_time_pk1 = [0,0,0,0]
    Alice_time_shared_key1 = [0,0,0,0]
    Bob_time_shared_key1 = [0,0,0,0]
    for level in range(0,4,1):

        eA = eA_list[level]
        eB = eB_list[level]
        p = gm.mul(pow(mpz(2),eA), pow(mpz(3),eB))-1


        ## The Elliptic Curve (The initial elliptic curve)
        ## Points on the Elliptic curve
        PA, QA, PB, QB,a,b = get_basis_points_and_curve_coefficients(level)
        PA = initit_point_affine(PA)
        QA = initit_point_affine(QA)
        PB = initit_point_affine(PB)
        QB = initit_point_affine(QB)
        a = initit_fp2(a)
        b = initit_fp2(b)
        AlicePoints = [PA, QA]
        BobPoints = [PB, QB]

        ############################# Public keys part ###########################

        ## Public key of Alice
        ##------------------------------------------------------------
        rs = gm.random_state(hash(gm.random_state()))
        sk2 = gm.mpz_random(rs, mpz(2^eA))
        t = time.time()
        a_Alice_pk, b_Alice_pk,BobPointsfromA = isogen2(sk2)
        elapsed = time.time()-t
        print("Elapsed time in seconds to generate Alice pk: ", elapsed)
        Alice_time_pk1[level]+= elapsed
        del t


        ## Public key of Bob
        ##------------------------------------------------------------
        rs = gm.random_state(hash(gm.random_state()))
        pow1 =mpz(gm.floor(gm.log2(pow(gm.mpz(3),eB))))
        sk3 = gm.mpz_random(rs, mpz(2^pow1))
        t = time.time()
        a_Bob_pk, b_Bob_pk,AlicePointsfromB= isogen3(sk3)
        elapsed = time.time()-t
        print("Elapsed time in seconds to generate Bob pk: ", elapsed)
        Bob_time_pk1[level]+=elapsed
        del t

        ###############################  Shared key part #############################
        ## Alice Shaerd key
        ##-------------------------------------------------------------------------
        t = time.time()
        Alice_shared_key = isox2(sk2,a_Bob_pk,b_Bob_pk,AlicePointsfromB)
        elapsed = time.time()-t
        print("Elapsed time to get the shared key of Alice is: ", elapsed)
        Alice_time_shared_key1[level]+=elapsed
        del t

        ##Bob Shared key
        ##---------------------------------------------------------------------------
        #print(sk3)
        t = time.time()
        Bob_shared_key = isox3(sk3,a_Alice_pk,b_Alice_pk,BobPointsfromA)
        elapsed = time.time()-t
        print("Elapsed time to get the shared key of Bob is: ", elapsed)
        Bob_time_shared_key1[level]+=elapsed
        del t

        ##print the shared key##
        if (Alice_shared_key == Bob_shared_key):
            print("The shared key is: ")
            print(int(Alice_shared_key[0])+int(Alice_shared_key[1])*i)

        else:
            print("some error happened in calculations ...shared keys aren't equals !!! :(")

        print("---------------------------------------------------------------------------")
        del eA, eB, p, PA, QA, PB, QB
    
    print('Iteration {} executed successfuly !!\n \n'.format(index))
    if(index == 0):
        Alice_time_pk1_multiple = np.array([[Alice_time_pk1]])
        Bob_time_pk1_multiple = np.array([[Bob_time_pk1]])
        Alice_time_shared_key1_multiple = np.array([[Alice_time_shared_key1]])
        Bob_time_shared_key1_multiple = np.array([[Bob_time_shared_key1]])
    else:
        row = np.array([Alice_time_pk1])
        Alice_time_pk1_multiple = np.r_[Alice_time_pk1_multiple,[row]]
        
        row = np.array([Bob_time_pk1])
        Bob_time_pk1_multiple = np.r_[Bob_time_pk1_multiple,[row]]
        
        row = np.array([Alice_time_shared_key1])
        Alice_time_shared_key1_multiple = np.r_[Alice_time_shared_key1_multiple,[row]]
        
        row = np.array([Bob_time_shared_key1])
        Bob_time_shared_key1_multiple = np.r_[Bob_time_shared_key1_multiple,[row]]

```

```python
file = open('time_measurement.txt','w')
for index in range(0, iteration):
    Alicetimepk1 = str(Alice_time_pk1_multiple[index][0].tolist())
    file = open('time_measurement.txt','a')
    file.write(Alicetimepk1+'\n')
    #-------------------------------------------
    Bobtimepk1 = str(Bob_time_pk1_multiple[index][0].tolist())
    file.write(Bobtimepk1+'\n')
    #--------------------------------------------
    Alicetimesharedkey1 = str(Alice_time_shared_key1_multiple[index][0].tolist())
    file.write(Alicetimesharedkey1+'\n')
    #-------------------------------------------
    Bobtimesharedkey1 = str(Bob_time_shared_key1_multiple[index][0].tolist())
    file.write(Bobtimesharedkey1+'\n')
    #-------------------------------------------
    file.close()
```
