```python
def Mont2wer(A):
    a = 1+0*i - A^2/3
    b = (2*A^3 - 9*A)/(27+0*i)
    return a,b
```

```python
def isogeny_2_eA(E_0, S_A, BobPoints):
    #The function computes the isogeny map of Alice of degree 2^{e_A} by decompossing the map into e_A isogeny map of degree 2
    # Input: -The initital elliptic curve E_0
    #        -Alice secret_point S_A = P_A+ [n_A].Q_A (with order 2^{e_A})
    #        -Bobpoints = [PB, QB]
    # Output: -Alice's public key consists of: E_A, the evaluation of Bob's points according to Alice's map [P_B_prime, Q_B_prime ]
    P_B = BobPoints[0]
    Q_B = BobPoints[1]
    E_0_prime = E_0
    S_A_prime = S_A
    P_B_prime = P_B
    Q_B_prime = Q_B   
    for i in range(1,eA+1):
        pow = eA -i
        R_A = 2^pow * S_A_prime
       # print("order 2", R_A)
        phi = E_0_prime.isogeny(R_A)
        E_0_prime = phi.codomain()
        S_A_prime = phi(S_A_prime)
        P_B_prime = phi(P_B_prime)
        Q_B_prime= phi(Q_B_prime) 
        
    return [phi.codomain(), [P_B_prime, Q_B_prime]]
```

```python
def isogeny_3_eB(E_0,S_B, AlicePoints):
    #The function computes the isogeny map of Bob with degree 2^{e_B} by decompossing the map into e_B isogeny map of degree 3
    # Input: - The initital elliptic curve E_0
    #        - Bob secret_point S_B = P_B+ [m_A].Q_B (with order 2^{e_B})
    #        - Alice's points = [PA, QA]
    # Output:  -Bob's public key consists of: E_B, the evaluation of Alice's points according to Bob's map [P_A_prime, Q_A_prime ]
    P_A = AlicePoints[0]
    Q_A = AlicePoints[1]
    E_0_prime = E_0
    S_B_prime = S_B
    P_A_prime = P_A
    Q_A_prime = Q_A
    for i in range(1,eB+1):
        pow = eB -i
        R_B = 3^pow * S_B_prime
        #print("order 3", R_B)
        phi = E_0_prime.isogeny(R_B)
        E_0_prime= phi.codomain()
        S_B_prime = phi(S_B_prime)
        P_A_prime = phi(P_A_prime)
        Q_A_prime = phi(Q_A_prime)
        #print(phi) 
        #print("R_bar", R_bar)
        #print("P_bar", Pa_bar)
        #print("Q_bar", Qa_bar)
    return [phi.codomain(), [P_A_prime, Q_A_prime]]
```

```python
def Alice_sk(E_B, S_A_prime):
    # This function computes the shared key of Alice
    #input: - Bob's public curve E_B
    #       - Alice's secret point S_A_prime calculated based on the received points that have beeen evaluated according to-
    #       - -Bob's map i.e S_A_prime = P_A_prime + [n_A].Q_A_prime
    # output: the shared key of Alice calculated as the j_invariant value of E_AB      
    E_AB = E_B
    #R_bar = R
    for i in range(1,eA+1):
        pow = eA -i
        R_A = 2^pow * S_A_prime
        #print("The point of order 2", R_A)
        phi = E_AB.isogeny(R_A)
        E_AB= phi.codomain()
        #print("E",Eab)
        S_A_prime = phi(S_A_prime)
       
    
    return phi.codomain().j_invariant()
```

```python
def Bob_sk(E_A, S_B_prime):
    # This function computes the shared key of Bob
    #input: - Alice's public curve E_A
    #       - Bob's secret point S_B_prime calculated based on the received points that have beeen evaluated according to-
    #       - -Alice's map i.e S_B_prime = P_B_prime + [m_A].Q_B_prime
    # output: the shared key of Bob calculated as the j_invariant value of E_BA       
    E_BA = E_A
    #R_bar = R
    for i in range(1,eB+1):
        pow = eB -i
        R_B = 3^pow * S_B_prime
        #print("point with order 3", R_B)
        phi = E_BA.isogeny(R_B)
        E_BA= phi.codomain()
        S_B_prime = phi(S_B_prime)
    return phi.codomain().j_invariant()
```

## Real World short-Weierstrass example (Measuring time and cycles needed to generate PK & get the shared key)

## 1. Measuring Time

```python
## Public Parameters: 
eA_list = [ 216, 250, 305, 372]
eB_list = [137, 159, 192,239]
prime = [434, 503, 610,751]
Alice_time_pk0 = [0,0,0,0]
Bob_time_pk0 = [0,0,0,0]
Alice_time_shared_key0 = [0,0,0,0]
Bob_time_shared_key0 = [0,0,0,0]


for level in range(0,4,1):
    
    eA = eA_list[level]
    eB = eB_list[level]
    p = 2^eA*3^eB - 1
    _.<I> = GF(p)[]
    K.<i> = GF(p^2, modulus=I^2+1)
    A = 6+0*i
    a,b = Mont2wer(A)
    E_0 = EllipticCurve(K,[a,b])
    
    P_A, Q_A, P_B, Q_B = get_basis_points(E_0,level)

    AlicePoints = [P_A, Q_A]
    BobPoints = [P_B, Q_B]

    ## Alice Pk generating 
    
    n_A = randint(0, 2^eA-1)
    #Sa =11
    t = time.time()
    S_A = AlicePoints[0] + n_A * AlicePoints[1]
    E_A, BobPointsfromA = isogeny_2_eA(E_0, S_A, BobPoints)
    elapsed = time.time()-t
    print("Elapsed time in seconds to generate Alice pk: ", elapsed)
    Alice_time_pk0[level]+= elapsed
    del t

    ## Bob pk generating
    n_B = randint(0, 3^eB-1)
    #Sb =2
    t = time.time()
    S_B = BobPoints[0] + n_B * BobPoints[1]
    E_B, AlicePointsfromB = isogeny_3_eB(E_0, S_B,AlicePoints)
    elapsed = time.time()-t
    print("Elapsed time in seconds to generate Bob pk: ", elapsed)
    Bob_time_pk0[level]+=elapsed
    del t

    ##Alice Shared key
    t = time.time()
    S_A_prime = AlicePointsfromB[0] + n_A*AlicePointsfromB[1]
    Alice_shared_key = Alice_sk(E_B,S_A_prime)
    elapsed = time.time()-t
    print("Elapsed time to get the shared key of Alice is: ", elapsed)
    Alice_time_shared_key0[level]+=elapsed
    del t

    ##Bob shared key
    t = time.time()
    S_B_prime = BobPointsfromA[0] + n_B*BobPointsfromA[1]
    Bob_shared_key = Bob_sk(E_A,S_B_prime)
    elapsed = time.time()-t
    print("Elapsed time to get the shared key of Bob is: ", elapsed)
    Bob_time_shared_key0[level]+=elapsed
    del t

    if (Alice_shared_key == Bob_shared_key):
        print("The shared key is: ")
        print(Alice_shared_key)

    else:
        print("some error happened in calculations ...shared keys aren't equals !!! :(")
        
    print("---------------------------------------------------------------------------")
    del eA, eB, p, P_A, Q_A, P_B, Q_B, K, I 
```

## 2.measuring the CPU Cycles 

```python
## Public Parameters: 
eA_list = [ 216, 250, 305, 372]
eB_list = [137, 159, 192,239]
prime = [434, 503, 610,751]
Alice_cpu_pk0 = [0,0,0,0]
Bob_cpu_pk0 = [0,0,0,0]
Alice_cpu_shared_key0 = [0,0,0,0]
Bob_cpu_shared_key0 = [0,0,0,0]


for level in range(0,4,1):
    
    eA = eA_list[level]
    eB = eB_list[level]
    p = 2^eA*3^eB - 1
    _.<I> = GF(p)[]
    K.<i> = GF(p^2, modulus=I^2+1)
    A = 6+0*i
    a,b = Mont2wer(A)
    E_0 = EllipticCurve(K,[a,b])
    
    P_A, Q_A, P_B, Q_B = get_basis_points(E_0,level)

    AlicePoints = [P_A, Q_A]
    BobPoints = [P_B, Q_B]

    ## Alice Pk generating 
    n_A = randint(0, 2^eA-1)
        #n_A =11
    with Timer() as alice_pk_timer:
        S_A = AlicePoints[0] + n_A * AlicePoints[1]
        E_A, BobPointsfromA = isogeny_2_eA(E_0, S_A, BobPoints)
    elapsed = alice_pk_timer.cycles
    print("Elapsed cpu cycles to generate Alice pk: ", elapsed)
    Alice_cpu_pk0[level]+= elapsed

    ## Bob pk generating
    n_B = randint(0, 3^eB-1)
        #n_B =2
    with Timer() as bob_pk_timer:
        S_B = BobPoints[0] + n_B * BobPoints[1]
        E_B, AlicePointsfromB = isogeny_3_eB(E_0, S_B,AlicePoints)
    elapsed = bob_pk_timer.cycles
    print("Elapsed cpu cycles to generate Bob pk: ", elapsed)
    Bob_cpu_pk0[level]+=elapsed
       

    ##Alice Shared key
    with Timer() as alice_sk_timer:
        S_A_prime = AlicePointsfromB[0] + n_A*AlicePointsfromB[1]
        Alice_shared_key = Alice_sk(E_B,S_A_prime)
    elapsed = alice_sk_timer.cycles
    print("Elapsed time to get the shared key of Alice is: ", elapsed)
    Alice_cpu_shared_key0[level]+=elapsed

    ##Bob shared key
    with Timer() as bob_sk_timer: 
        S_B_prime = BobPointsfromA[0] + n_B*BobPointsfromA[1]
        Bob_shared_key = Bob_sk(E_A,S_B_prime)
    elapsed = bob_sk_timer.cycles
    print("Elapsed time to get the shared key of Bob is: ", elapsed)
    Bob_cpu_shared_key0[level]+=elapsed

    if (Alice_shared_key == Bob_shared_key):
        print("The shared key is: ")
        print(Alice_shared_key)

    else:
        print("some error happened in calculations ...shared keys aren't equals !!! :(")
        
    print("---------------------------------------------------------------------------")
    del eA, eB, p, P_A, Q_A, P_B, Q_B, K, I 
```
